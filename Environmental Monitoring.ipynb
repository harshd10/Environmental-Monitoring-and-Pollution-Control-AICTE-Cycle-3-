```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load dataset

df = pd.read_csv("city_day.csv")
df.head()
```

```python
df.info()
```

```python
#Missing values per column
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
# Histograms for all numeric columns
df.hist(figsize=(12,10), bins=20)
plt.suptitle("Distribution of Numeric Features")
plt.show()
```

```python
#Group data by City and calculate the mean of pollutants
city_mean_df = df.groupby('City')[['PM2.5', 'PM10', 'NO2', 'SO2']].mean().sort_values(by='PM2.5', ascending=False)

#Display the average pollutant levels per city
print("Average pollutant levels per city:")
print(city_mean_df)

#Visualization the average PM2.5 levels per city
plt.figure(figsize=(15, 8))
sns.barplot(x=city_mean_df.index, y=city_mean_df['PM2.5'], palette='viridis')
plt.title('Average PM2.5 levels Across different cities')
plt.xlabel('City')
plt.ylabel('Average PM2.5 Concentration')
plt.xticks(rotation=45)
plt.show()
```

```python
#Select a few major cities for comparison
major_cities = ['Delhi', 'Mumbai', 'Bengaluru', 'Chennai']
subset_df = df[df['City'].isin(major_cities)]

#Create a box plot to compare PM2.5 distribution across cities
plt.figure(figsize=(12, 8))
sns.boxplot(x='City', y='PM2.5', data=subset_df, palette='Set2')
plt.title('Distribution of PM2.5 Levels in Major Cities')
plt.xlabel('City')
plt.ylabel('PM2.5 Concentration')
plt.show()
```

```python
#Select only numeric columns
numeric_df = df.select_dtypes(include='number')

#Plot correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatap of Numeric Features")
plt.show()
```
